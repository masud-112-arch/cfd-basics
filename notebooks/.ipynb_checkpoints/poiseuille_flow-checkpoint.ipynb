# 2D Poiseuille Flow (Pressure-Driven Channel Flow)

Goal: Solve steady 2D incompressible flow between two parallel plates under a constant pressure gradient.

---

## Governing Equations

For steady, laminar, fully developed flow in x-direction:

$$
\frac{\partial^2 u}{\partial y^2} = \frac{1}{\mu} \frac{dP}{dx}
$$

- $u(y)$ = velocity along x (function of y only)
- $\mu$ = dynamic viscosity
- $\frac{dP}{dx}$ = constant pressure gradient driving the flow
- Boundary conditions: $u(0) = 0$, $u(H) = 0$ (no-slip at walls)

---

## Analytical Solution

$$
u(y) = \frac{1}{2\mu} \frac{dP}{dx} \, y(H-y)
$$

- Parabolic profile between two walls.

---

## Numerical Approach

- Discretize y using finite difference (central difference):
$$
\frac{u_{i+1} - 2 u_i + u_{i-1}}{\Delta y^2} = \frac{1}{\mu} \frac{dP}{dx}
$$
- Solve linear system for \(u_i\) numerically.
- Compare numerical and analytical solution.

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# Parameters
# ---------------------------
H = 1.0           # channel height
Ny = 50           # number of grid points
mu = 1.0          # dynamic viscosity
dPdx = -1.0       # pressure gradient (negative for flow in +x direction)

dy = H / (Ny - 1)
y = np.linspace(0, H, Ny)

# ---------------------------
# Analytical solution
# ---------------------------
u_analytical = (1/(2*mu)) * dPdx * y * (y - H)

# ---------------------------
# Numerical solution using central difference
# ---------------------------
# Set up tridiagonal system Au = b
A = np.zeros((Ny, Ny))
b = np.zeros(Ny)

for i in range(Ny):
    if i == 0 or i == Ny-1:
        # boundary condition u=0 at walls
        A[i,i] = 1.0
        b[i] = 0.0
    else:
        A[i,i-1] = 1.0 / dy**2
        A[i,i]   = -2.0 / dy**2
        A[i,i+1] = 1.0 / dy**2
        b[i] = dPdx / mu

# Solve linear system
u_numerical = np.linalg.solve(A, b)

# ---------------------------
# Plot results
# ---------------------------
plt.figure(figsize=(7,5))
plt.plot(u_analytical, y, 'r-', lw=2, label='Analytical')
plt.plot(u_numerical, y, 'bo', label='Numerical', markersize=4)
plt.xlabel('Velocity u(y)')
plt.ylabel('y')
plt.title('2D Poiseuille Flow Between Parallel Plates')
plt.legend()
plt.grid(True)
plt.show()
```

## Observations

- Numerical solution matches the analytical parabolic profile perfectly.
- Central difference approximation is second-order accurate in space.
- This simple test case is widely used to validate CFD codes.
